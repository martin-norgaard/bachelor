```python
import pandas as pd
```

```python
#concacenate csv-file
file_names = [f'data/AMECO{i}.csv' for i in range(1,13)]
combined_csv = pd.concat([pd.read_csv(f, sep=";") for f in file_names])
combined_csv.to_csv('data/AMECO.csv', sep=";", index = False)
```

```python
df = combined_csv.drop(columns=['COUNTRY','SUB-CHAPTER','TITLE','UNIT'])
df[["iso","code"]] = df.CODE.str.split(".",n=1,expand = True)
df = df.drop(columns=["CODE","Unnamed: 67"])
```

```python
countries = ['AUS','BEL','CAN','DNK','FIN','FRA','DEU','ITA','JPN','NLD','NOR','PAT','CHE','GBR','USA']
codes = ['1.0.0.0.NPTD','3.1.0.0.PIGT','3.0.0.0.ZCPIN','1.0.0.0.NETD','3.0.0.0.ZVGDF','1.0.0.0.ALCD0','1.1.0.0.OIGT','1.0.0.0.OKND','1.0.0.0.AKNDV','1.3.0.0.ISRC']
df = df[df['iso'].isin(countries)]
df = df[df['code'].isin(codes)]
df = df.melt(id_vars=['iso','code'], 
        var_name='year', 
        value_name="value")
#df = df.set_index('year')
df
```
